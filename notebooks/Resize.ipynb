```python
import numpy
from PIL import Image

imarray = numpy.random.rand(200,50,3) * 255
im = Image.fromarray(imarray.astype('uint8')).convert('RGBA')
```

```python
z = transforms.PILToTensor()(im).unsqueeze(0)
```

```python
rs_b = transforms.Resize(z.shape[-2:], interpolation = InterpolationMode.BICUBIC)
rs_n = transforms.Resize(z.shape[-2:], interpolation = InterpolationMode.NEAREST)
rs_bl = transforms.Resize(z.shape[-2:], interpolation = InterpolationMode.BILINEAR)
s = transforms.Resize(128)
```
